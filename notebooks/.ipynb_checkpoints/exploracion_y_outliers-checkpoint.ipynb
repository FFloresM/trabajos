# c.
outliers detection: Mahalanobis Distance

```python
#usaremos solo algunas características del dataset
datos_red = datos[['a','e','i','w','q','ad']].dropna()
datos_red.head()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
# Define el objeto PCA (principal component analisys)
pca = PCA()
# Ejecuta PCA con datos escalados y obtiene el arreglo de puntos
T = pca.fit_transform(StandardScaler().fit_transform(datos_red))
# _Grafico de las primeras 2 Dimensiones 
fig = plt.figure(figsize=(8,6))
with plt.style.context(('ggplot')):
    plt.scatter(T[:, 1], T[:, 0], edgecolors='k', cmap='jet')
    plt.xlabel('PC1')
    plt.ylabel('PC2')
plt.show()
```

```python
T
```

```python
from sklearn.covariance import EmpiricalCovariance, MinCovDet

# ajusta un estimador Minimum Covariance Determinant (MCD) a los datos 
robust_cov = MinCovDet().fit(T[:,:5])

# obtiene la distancia Mahalanobis
m = robust_cov.mahalanobis(T[:,:5])
```

```python
#colores para determinar los outliers
max_m = max(m)
colors = [plt.cm.jet(float(i)/max_m) for i in m]
```

```python
fig = plt.figure(figsize=(8,6))
with plt.style.context(('ggplot')):
    plt.scatter(T[:, 1], T[:, 0], c=colors, edgecolors='k', s=60)
    plt.xlabel('PC1')
    plt.ylabel('PC2')
    plt.title('Score Plot')
plt.show()
```

```python
puntos = np.array(colors)
```

```python
puntos
```

```python
#obtener los valores punicos, indices y cantidad 
u, indices, counts = np.unique(puntos, axis=0, return_index=True, return_counts=True)
```
